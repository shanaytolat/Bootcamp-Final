# **Exploratory data analysis (EDA)**

```python
import pandas as pd
df = pd.read_csv("dataset.csv") #Load the dataset

import seaborn as sns
import matplotlib.pyplot as plt

#Scatterplot with regression line: Average leaf area per plant (ALAP) vs PDMVG
plt.figure(figsize=(8, 5))
sns.regplot(x='ALAP', y='PDMVG', data=df, scatter_kws={'alpha': 0.5, 'color': 'green'}, line_kws={'color': 'green'})
plt.title("ALAP vs PDMVG")
plt.xlabel("Average leaf area per plant (ALAP)")
plt.ylabel("PDMVG)")
plt.grid(True)
plt.tight_layout()
plt.show()

#Scatterplot with regression line: Average number of leaves per plants (ANPL) vs PDMVG
plt.figure(figsize=(8, 5))
sns.regplot(x='ANPL', y='PDMVG', data=df, scatter_kws={'alpha': 0.5, 'color': 'green'}, line_kws={'color': 'green'})
plt.title("ANPL vs PDMVG")
plt.xlabel("Average number of leaves per Plant")
plt.ylabel("PDMVG")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Split PDMVG into quartiles
df['PDMVG_q'] = pd.qcut(df['PDMVG'], q=4, labels=["Q1", "Q2", "Q3", "Q4"])

#Create a violin plot of Average Chlorophyll Content (ACHP) versus PDMVG
plt.figure(figsize=(8, 6))
sns.violinplot(x='PDMVG_q', y='ACHP', data=df, palette='Greens')
plt.title("ACHP distribution across PDMVG Quartiles")
plt.xlabel("PDMVG Quartiles")
plt.ylabel("ACHP")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Create a boxplot to show distribution of plant height growth rate and PDMVG quartiles
plt.figure(figsize=(8, 6))
sns.boxplot(x='PDMVG_q', y='PHR', data=df, palette='Greens')
plt.title("PHR distribution across PDMVG quartiles")
plt.xlabel("PDMVG quartile")
plt.ylabel("PHR")
plt.tight_layout()
plt.show()
```

```python
#Distribution of PDMVG across the dataset range
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='PDMVG', bins=20, kde=True, color='green')
plt.title('Distribution of PDMVG')
plt.xlabel('PDMVG (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Correlation heat map of root-related features and PDMVG
root_features = ['ARD', 'ARL', 'ADWR', 'AWWR', 'PDMVG']
corr_matrix = df[root_features].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='Greens', fmt=".2f", square=True, linewidths=0.5)
plt.title('Correlation heatmap of root-related features')
plt.show()
```
